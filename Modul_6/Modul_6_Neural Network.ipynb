#### Konsep Dasar Bobot dan Pengaruh Sinapsis

Ide utamanya adalah bahwa kekuatan sinaptik (bobot w) dapat dipelajari dan mengontrol kekuatan pengaruh serta arah pengaruhnya: eksitatori (bobot positif) atau inhibitor (bobot negatif) dari satu neuron ke neuron lainnya. Dalam model dasar, dendrit membawa sinyal ke badan sel di mana semuanya dijumlahkan. Jika jumlah akhir melebihi ambang batas tertentu, neuron dapat menembak, mengirimkan lonjakan sepanjang aksonnya.

Dalam model komputasi, diasumsikan bahwa waktu tepat dari lonjakan tidak penting, dan hanya frekuensi penembakan yang mengkomunikasikan informasi. Kita memodelkan tingkat penembakan neuron dengan fungsi aktivasi (misalnya, fungsi sigmoid), yang mewakili frekuensi lonjakan sepanjang akson.

## Fungsi Aktivasi

Activation function adalah sebuah persamaan yang digunakan untuk mengubah input dari sebuah node dalam sebuah lapisan menjadi sebuah output. Fungsi ini adalah semacam gerbang logika di antara input untuk neuron saat ini dan outputnya menuju ke neuron berikutnya.

### Contoh  

<b>TanH</b>

```python
def tanh(z):
    return np.tanh(z)
```

<b>Rectified Linear Unit (ReLU)</b>

```python
def relu(z):
    return max(0, z)
```

<b>Softmax</b>

```python
def softmax(x):
    ex = np.exp(x)
    sum_ex = np.sum(np.exp(x))
    return ex / sum_ex
```

Selain fungsi aktivasi yang disebutkan diatas, terdapat banyak fungsi aktivasi lainnya seperti yang ada dibawah ini.

## Propagation

### Forward Propagation

Forward propagation adalah proses memasukkan input dari arah depan ke dalam neural network. Tiap layer akan menerima input, kemudian memprosesnya sesuai fungsi aktivasi dan meneruskannya ke layer berikutnya.

## Learning Rate

Learning rate menentukan besarnya langkah yang diambil model selama gradient descent, yang merupakan teknik yang digunakan untuk meminimalkan loss function. Memilih learning rate yang tepat sangat penting, karena learning rate yang tinggi memungkinkan model untuk belajar dengan cepat dengan mengambil langkah yang lebih besar, tetapi berisiko melewati minimum loss. Di sisi lain, learning rate yang rendah memastikan kemajuan yang pasti sedikit demi sedikit, namun berpotensi menyebabkan model terjebak dalam local minima.
