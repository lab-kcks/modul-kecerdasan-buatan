# Modul 4: Supervised Learning

Supervised learning adalah salah satu bidang machine learning di mana model belajar dari data pelatihan berlabel. Selama pelatihan, pengguna memberikan data input kepada algoritma serta label output yang terkait. Berdasarkan data input, algoritma mempelajari pemetaan dari input ke output dan menghasilkan prediksi atau penilaian. Ada banyak penerapan yang menggunakan supervised learning, termasuk penyaringan spam, pengenalan suara, penerjemahan mesin, dan periklanan online.

Dalam supervised learning, dua kategori utama yang akan dibahas pada modul ini adalah **Regresi** dan **Klasifikasi**. Regresi melibatkan prediksi nilai numerik yang kontinu, sedangkan klasifikasi melibatkan penempatan prediksi ke dalam kategori atau kelas yang diskrit. Namun, ada berbagai supervised learning lainnya yang memiliki tujuan dan aplikasi spesifiknya, seperti time series forecasting, sequence labeling, object detection, dan banyak lagi.

Sebelum memasuki bagian pembahasan lainnya, ada dua istilah yang perlu kita ketahui:

*   **Independent variables** adalah input dalam model supervised learning. Variabel-variabel ini adalah variabel yang dimanipulasi untuk mengamati efeknya terhadap variabel dependen.
    
    Contoh: Bayangkan Anda sedang membuat kue. Jumlah gula yang Anda masukkan adalah variabel independen - itulah yang dapat Anda ubah untuk membuat kue lebih manis atau kurang manis.

*   **Dependent Variables** adalah output dari model. Variabel ini adalah variabel yang ingin kita prediksi atau pahami.
    
    Contoh: Rasa manis dari kue adalah variabel dependen. Tergantung pada jumlah gula yang ditambahkan.

## Regresi
  
Regresi adalah salah satu jenis metode Supervised Learning di mana variabel targetnya berupa nilai yang bersifat kontinu. Contoh penggunaannya termasuk memprediksi berat badan, usia, harga, dan sebagainya. Terdapat berbagai algoritma yang digunakan dalam metode regresi ini yang selanjutnya akan kita bahas.

### Linear regression

Linear regression adalah metode statistik yang memodelkan hubungan antara dua variabel atau lebih dengan mencocokkan persamaan linear dengan data yang diamati.
Suatu cara untuk mengetahui hubungan antara variabel independen dan variabel dependen dalam sebuah garis lurus (seperti perosotan di taman bermain).


Contoh: Misalkan Anda menjalankan gerai minuman limun dan Anda melihat bahwa semakin panas di luar, semakin banyak limun yang Anda jual. Regresi linier membantu Anda mengetahui seberapa banyak limun yang akan Anda jual untuk setiap derajat lebih panas. Selain itu, misalkan kita ingin memprediksi berat badan seseorang berdasarkan tinggi badannya. Kita dapat menggunakan regresi linier untuk menemukan garis yang paling cocok yang menggambarkan hubungan antara tinggi dan berat badan.

#### Simple Linear Regression.

Simple Regression adalah metode statistik yang memperkirakan hubungan antara satu variabel dependen dan satu variabel independen. Ini adalah jenis model regresi yang digunakan untuk menggambarkan hubungan antara dua variabel kuantitatif. Simple Regression juga dikenal sebagai simple linear regression, karena melibatkan pencocokan garis lurus dengan data yang diamati.

Rumus untuk Simple Linear Regression adalah:

![image.png](Images/Simple%20Regression.png)

>- y = nilai prediksi variabel dependen (y) untuk setiap nilai variabel independen (x) yang diberikan.  
>- $\beta_{0}$ = intercept, nilai prediksi y ketika x = 0.  
>- $\beta_{1}$ = koefisien regresi, seberapa besar kita mengharapkan y berubah ketika x meningkat.  
>- x = variabel independen (variabel yang kita harapkan mempengaruhi y).  
>- $\epsilon$ = kesalahan estimasi, atau seberapa besar variasi yang ada dalam estimasi kita terhadap koefisien regresi.  

Contoh dari Simple Regression ini yakni kasus income dan happiness sebagai berikut:

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
import statsmodels.api as sm

# Load data
income_data = pd.read_csv('https://raw.githubusercontent.com/naadimpra/kebutuhanPraktikumAI/main/income.data.csv')
```

```python
income_data.head()
```

```python
# Definisikan Independent dan Dependent Variable
X = income_data['income']  # Independent variable
y = income_data['happiness']  # Dependent variable

# Menentukan starting point y bila x=0
X = sm.add_constant(X)

# Penggunaan model
model = sm.OLS(y, X).fit()
```

```python
print(model.summary())
```

Model ini telah digunakan untuk memahami hubungan antara kebahagiaan (variabel dependen) dan pendapatan (variabel independen).

>*  Dep. Variable: Hasil atau target yang coba diprediksi oleh model.
>*  R-squared: Statistik yang memberi tahu kita proporsi variasi dalam variabel dependen yang dijelaskan oleh model. Nilai yang lebih tinggi berarti kecocokan yang lebih baik.
>*  Adj. R-squared: Mirip dengan R-squared namun disesuaikan dengan jumlah prediktor dalam model; ini lebih dapat diandalkan ketika membandingkan model dengan jumlah prediktor yang berbeda.
>*  F-statistic: Ukuran signifikansi keseluruhan model. Nilai yang lebih tinggi menunjukkan model yang lebih signifikan secara statistik.
>*  Prob ( F-statistic): Probabilitas bahwa hasil yang diamati dapat terjadi jika model tidak memiliki kekuatan prediksi yang nyata; nilai yang lebih rendah menunjukkan model yang lebih berguna.
>*  Log-Likelihood: Ukuran seberapa baik model cocok dengan data; nilai yang lebih tinggi berarti kecocokan yang lebih baik.
>*  AIC/BIC: Ini adalah kriteria untuk pemilihan model, menggabungkan kecocokan model dan kompleksitas; nilai yang lebih rendah menunjukkan model yang lebih baik.
>*  Df Residual: Derajat kebebasan dari residual, yang merupakan jumlah observasi dikurangi jumlah parameter yang diestimasi.
>*  Df Model: Jumlah prediktor dalam model.
>*  Covariance Type: Menunjukkan asumsi tentang varians dari kesalahan; 'nonrobust' menunjukkan asumsi OLS standar.
>*  coef: Estimasi efek dari setiap prediktor. Ini memberi tahu kita seberapa besar variabel dependen diperkirakan akan berubah ketika prediktor berubah satu unit.
>*  std err: Kesalahan standar dari estimasi koefisien. Ini mengukur ketepatan estimasi koefisien.
>*  t: Statistik t untuk menguji hipotesis nol bahwa koefisien adalah nol. Nilai absolut yang besar menunjukkan bahwa prediktor tersebut adalah prediktor yang penting.
>*  P>|t|: Nilai p-value untuk uji hipotesis koefisien. Nilai yang rendah menunjukkan bahwa pengaruh prediktor signifikan secara statistik.
>*  [0.025 0.975]: Interval kepercayaan 95% untuk estimasi koefisien. Ini memberikan kisaran di mana kita dapat 95% yakin bahwa koefisien yang sebenarnya berada.


```python
# mendapatkan persamaan regresi
x_pred = np.linspace(X['income'].min(), X['income'].max(), 50)  # 50 spot representasi income yang rata dari terendah hingga tertinggi
x_pred2 = sm.add_constant(x_pred)  # menentukan base height dan starting point
y_pred = model.predict(x_pred2) # memprediksi data dengan model
```

```python
import matplotlib.pyplot as plt

# buat plot
plt.figure(figsize=(10, 6))
plt.scatter(X['income'], y, label='Data points')
plt.plot(x_pred, y_pred, color='red', label=f'Regression Line: \nHappiness = {model.params[0]:.2f} + {model.params[1]:.2f}*Income')
plt.xlabel('Income')
plt.ylabel('Happiness')
plt.title('Simple Linear Regression of Happiness on Income')
plt.legend()
plt.grid(True)
plt.show()
```

model.params[0] mengacu pada intersep model, yang merupakan nilai rata-rata yang diharapkan dari variabel dependen ketika semua variabel independen disetel ke nol.

model.params[1] mengacu pada koefisien variabel independen pertama dalam model, yang dalam hal ini adalah 'Income'. Koefisien ini merepresentasikan perubahan dalam variabel dependen 'Happiness' untuk perubahan satu unit dalam 'Income'.

Hasil penelitian menunjukkan hubungan yang signifikan (p <0,001) antara pendapatan dan kebahagiaan (R2 = 0,71  0,019), dengan peningkatan sebesar 0,71 unit dalam 'Happiness' yang dilaporkan untuk setiap 10.000 peningkatan 'Income'.

**Memeriksa Metrik Evaluasi**

```python
r_squared = model.rsquared
print(f'R-squared: {r_squared:.2f}')

mae = mean_absolute_error(y, model.predict(X))
print(f'Mean Absolute Error: {mae:.2f}')

mse = mean_squared_error(y, model.predict(X))
print(f'Mean Squared Error: {mse:.2f}')

rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse:.2f}')
```

*   "R-squared, atau koefisien determinasi, mengukur proporsi varians dalam variabel dependen yang dapat diprediksi dari variabel independen dalam model regresi; nilai yang mendekati 1 lebih baik karena berarti model tersebut menjelaskan sebagian besar varians, sementara nilai yang mendekati 0 lebih buruk karena menunjukkan model tersebut gagal menjelaskan sebagian besar varians."

*   "Mean Absolute Error (MAE) mengukur besarnya rata-rata kesalahan dalam satu set prediksi, tanpa mempertimbangkan arahnya; MAE yang lebih rendah lebih baik, mengindikasikan bahwa prediksi lebih dekat dengan nilai aktual, dan MAE yang lebih tinggi lebih buruk, yang berarti prediksi lebih melenceng secara rata-rata."

*   "Mean Squared Error (MSE) mewakili rata-rata kuadrat kesalahan, secara efektif menghukum kesalahan yang lebih besar lebih parah daripada MAE; dengan demikian, MSE yang lebih rendah lebih baik karena menunjukkan kecocokan yang lebih dekat dengan data, dan MSE yang lebih tinggi lebih buruk karena mengindikasikan kesalahan rata-rata yang lebih besar."

*   "Root Mean Squared Error (RMSE) adalah akar kuadrat dari MSE dan memberikan ukuran kesalahan antara nilai yang diamati dan yang diprediksi dalam unit yang sama dengan data; RMSE yang lebih rendah lebih baik karena mengindikasikan kesalahan yang lebih kecil, dan RMSE yang lebih tinggi lebih buruk, yang mencerminkan kesalahan yang lebih besar dengan penekanan yang lebih besar untuk menghukum ketidaksesuaian yang lebih besar."

#### Multiple Regression.

Multiple regression adalah teknik statistik yang memperkirakan hubungan antara variabel dependen dan dua atau lebih variabel independen. Metode ini merupakan lanjutan dari Simple Regression, yang hanya melibatkan satu variabel independen. Multiple regression juga dikenal sebagai multiple linear regression (MLR).

Rumus untuk regresi linier berganda adalah:

![image.png](Images/Multiple%20Regression.png)

>$y$ = nilai prediksi dari variabel dependen
>
>$\beta_{0}$ = intersep y (nilai y ketika semua parameter lain ditetapkan ke 0)
>
>$\beta_{1}X_{1}$ = koefisien regresi (B1) dari variabel independen pertama (X1) (alias efek peningkatan nilai variabel independen terhadap nilai y yang diprediksi)
>
>... = lakukan hal yang sama untuk berapapun banyak variabel independen yang Anda uji
>
>$\beta_{n}X_{n}$ = koefisien regresi dari variabel independen terakhir
>
>$\epsilon$ = kesalahan model (alias berapa banyak variasi yang ada dalam perkiraan kita tentang y)

Analisis ini memungkinkan kita untuk menilai seberapa baik kombinasi variabel bebas dapat memprediksi nilai variabel terikat dan seberapa signifikan masing-masing variabel bebas dalam membuat prediksi tersebut. Hal ini dilakukan dengan menghitung koefisien regresi yang meminimalkan kesalahan model, menggunakan statistik t untuk menilai keandalan model secara keseluruhan, dan nilai p untuk mengukur probabilitas bahwa hubungan yang diamati terjadi secara kebetulan.

Sebagai contoh praktis, dalam kasus restoran, regresi berganda dapat digunakan untuk menentukan bagaimana faktor-faktor seperti kualitas makanan, layanan, dan dekorasi mempengaruhi harga makanan. Dengan demikian, kita dapat memahami variabel mana yang paling mempengaruhi pembentukan harga dan seberapa kuat pengaruhnya.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

# Load the data
url = 'https://raw.githubusercontent.com/naadimpra/kebutuhanPraktikumAI/main/heart.data.csv'
heart_data = pd.read_csv(url)
```

```python
heart_data.head()
```

```python
# Menentukan Variabel
X = heart_data[['biking', 'smoking']]  # Independent variables
y = heart_data['heart.disease']  # Dependent variable

# Menentukan starting point y bila x=0
X = sm.add_constant(X)

# Fit the multiple linear model
model = sm.OLS(y, X).fit()
```

```python
# View the results
print(model.summary())
```

Dalam survei di 500 kota, ditemukan hubungan yang signifikan antara frekuensi bersepeda ke tempat kerja dan frekuensi penyakit jantung serta frekuensi merokok dan frekuensi penyakit jantung (p <0,001 untuk masing-masing). Secara khusus, ditemukan penurunan 0,2% ( 0,001) dalam frekuensi penyakit jantung untuk setiap 1% peningkatan bersepeda, dan peningkatan 0,178% ( 0,004) dalam frekuensi penyakit jantung untuk setiap 1% peningkatan merokok.

```python
# Hitung min max mean
min_smoking = heart_data['smoking'].min()
mean_smoking = heart_data['smoking'].mean()
max_smoking = heart_data['smoking'].max()

# Cetak value
print(f"Minimum smoking rate: {min_smoking}")
print(f"Mean smoking rate: {mean_smoking}")
print(f"Maximum smoking rate: {max_smoking}")
```

```python
# Tentukan rentang value bersepeda untuk sumbu x
biking_percentages = np.linspace(heart_data['biking'].min(), heart_data['biking'].max(), 100)

# Buat gambar dan sumbu
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot untuk data asli
ax.scatter(heart_data['biking'], heart_data['heart.disease'], alpha=0.5, label='Data Asli')

# Gambarkan garis untuk setiap value merokok
for smoking_percentage in [0.53, 15.44, 29.95]:
    # Buat DataFrame dengan value bersepeda dan value merokok yang konstan
    input_data = pd.DataFrame({'const': 1, 'biking': biking_percentages, 'smoking': smoking_percentage})
    # Pastikan kolom-kolom dalam urutan yang benar
    input_data = input_data[['const', 'biking', 'smoking']]
    # Dapatkan value penyakit jantung yang diprediksi
    predicted_heart_disease = model.predict(input_data)
    ax.plot(biking_percentages, predicted_heart_disease, label=f'Merokok = {smoking_percentage}')

# Beri label pada sumbu dan tambahkan legenda
ax.set_xlabel('Bersepeda ke Tempat Kerja (%)')
ax.set_ylabel('Persentase Penyakit Jantung dalam Populasi')
ax.legend()
ax.grid(True)
ax.set_title('Prediksi Persentase Penyakit Jantung vs Persentase Bersepeda ke Tempat Kerja')

# Tampilkan plot
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Get the predicted values
predicted_values = model.predict(X)

# Evaluation Metrics
r_squared = model.rsquared
print(f'R-squared: {r_squared:.2f}')

mae = mean_absolute_error(y, predicted_values)
print(f'Mean Absolute Error: {mae:.2f}')

mse = mean_squared_error(y, predicted_values)
print(f'Mean Squared Error: {mse:.2f}')

rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse:.2f}')
```

### Non Linear Regression.

Nonlinear regression adalah metode statistik yang digunakan untuk memodelkan hubungan antara variabel dependen dengan satu atau lebih variabel independen ketika hubungan tersebut tidak linier. Dengan kata lain, metode ini digunakan untuk menyesuaikan kurva dengan data ketika hubungan antara variabel tidak berbentuk garis lurus. Nonlinear regression digunakan di berbagai bidang, termasuk epidemiologi, keuangan, dan teknik.
Model regresi nonlinier dapat memiliki banyak bentuk, termasuk di antaranya adalah fungsi polinomial, eksponensial, dan logaritmik. Pilihan model tergantung pada sifat data dan pertanyaan penelitian yang diajukan. Model nonlinear regression diperkirakan menggunakan algoritma berulang yang meminimalkan perbedaan antara data yang diamati dan nilai prediksi dari model.

Contohnya Memprediksi probabilitas risiko kanker: Nonlinear regression dapat digunakan untuk mempelajari kemungkinan hubungan antara risiko kanker dan risiko penyakit kardiovaskular/stroke selama periode 12 tahun untuk pasien diabetes tipe 2. Dengan menggunakan model analisis regresi, probabilitas risiko kanker yang diprediksi dapat diperkirakan berdasarkan berbagai kondisi medis dan detail gaya hidup yang dikumpulkan datanya.

```python
import pandas as pd
from sklearn.metrics import r2_score
url='https://raw.githubusercontent.com/naadimpra/kebutuhanPraktikumAI/main/china_gdp.csv'
df = pd.read_csv(url)
df.head()
```

**Memplot Dataset**

Seperti inilah tampilan data poin. Terlihat seperti fungsi logistik atau eksponensial. Pertumbuhannya dimulai dengan lambat, kemudian dari tahun 2005 dan seterusnya, pertumbuhannya sangat signifikan. Dan akhirnya, pertumbuhannya sedikit melambat di tahun 2010-an.

```python
plt.figure(figsize=(8,5))
x_data, y_data = (df["Year"].values, df["Value"].values)
plt.plot(x_data, y_data, 'ro')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

**Memilih model**

Dari pengamatan awal pada plot, kita menentukan bahwa fungsi logistik dapat menjadi perkiraan yang baik, karena fungsi ini memiliki sifat dimulai dengan pertumbuhan yang lambat, pertumbuhan yang meningkat di tengah, dan kemudian menurun lagi di bagian akhir; seperti yang diilustrasikan di bawah ini:

`X = np.arange(-5,5.0, 0.1)`: Langkah ini membuat array `X` menggunakan fungsi `arange` dari NumPy, yang menghasilkan nilai dari -5 sampai di bawah 5 (karena nilai berhenti bersifat eksklusif) dengan langkah 0,1. Array ini akan berfungsi sebagai "variabel independen" untuk plot fungsi sigmoid.

`Y = 1.0 / (1.0 + np.exp(-X))`: Di sini kita menghitung fungsi sigmoid. `np.exp(-X)` menghitung eksponensial dari `-X` untuk setiap elemen dalam array `X`. Hasilnya kemudian digunakan dalam rumus fungsi sigmoid untuk mendapatkan nilai `Y` yang sesuai, yang berfungsi sebagai "variabel dependen".

```python
X = np.arange(-5,5.0, 0.1)
Y = 1.0 / (1.0 + np.exp(-X))

plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

Fungsi sigmoid yang didefinisikan di sini mewakili kurva sigmoid parametrik, yang merupakan jenis fungsi logistik. Fungsi ini sering digunakan dalam regresi logistik dan neural network karena fungsi ini dapat memetakan nilai input apa pun ke rentang antara 0 dan 1, sehingga sangat berguna untuk memodelkan distribusi probabilitas dan klasifikasi biner.


`y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2))))`

Di dalam fungsi tersebut, variabel y dihitung sebagai nilai fungsi

*   sigmoid pada x dengan parameter yang diberikan Beta_1 dan Beta_2.
np.exp adalah fungsi eksponensial e^x, yang merupakan fungsi matematika dasar yang sering digunakan dalam representasi alami dari proses pertumbuhan atau penyusutan.

*   Ekspresi `-Beta_1*(x-Beta_2)` menskalakan dan menggeser input x. Ketika Beta_1 positif, fungsi ini akan menghasilkan kurva berbentuk S standar; ketika Beta_1 negatif, kurva dibalik secara horizontal.

*   `1 + np.exp(-Beta_1*(x-Beta_2))` menggeser fungsi eksponensial secara vertikal sebesar 1 untuk memastikan output berkisar antara 0 dan 1.

**Pembuatan Model**

Sekarang, mari kita buat model regresi kita dan menginisialisasi parameternya.

```python
def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y
```

**Scale value agar dapat merepresentasikan GDP aktual.**

Hal ini dikarenakan output dari fungsi sigmoid berkisar antara 0 dan 1, dan kita perlu menskalakannya ke nilai PDB aktual untuk membuat perbandingan yang berarti. Garis yang diplot menunjukkan prediksi GDP menurut model dengan tebakan parameter awal.
Mari kita lihat contoh sigmoid line yang mungkin sesuai dengan data:

*   beta_1: Angka ini menunjukkan seberapa curam kurva Anda.
*   beta_2: Angka ini memberi tahu Anda di mana tikungan tengah kurva pada sumbu X (tahun).

```python
beta_1 = 0.10
beta_2 = 1990.0

#logistic function
Y_pred = sigmoid(x_data, beta_1 , beta_2)

#plot initial prediction against datapoints
plt.plot(x_data, Y_pred*15000000000000.)
plt.plot(x_data, y_data, 'ro')
```

Tugas kita di sini adalah menemukan parameter terbaik untuk model kita. Pertama-tama, mari kita normalisasi x dan y kita:

```python
# Normalisasi Data
xdata =x_data/max(x_data)
ydata =y_data/max(y_data)
```

**Bagaimana kita menemukan parameter terbaik untuk garis fit kita?**

Kita dapat menggunakan curve_fit yang menggunakan kuadrat terkecil non-linear untuk menyesuaikan fungsi sigmoid kita, dengan data. Nilai-nilai optimal untuk parameter-parameternya sehingga jumlah kuadrat residu sigmoid(xdata, *popt) - ydata dapat diminimalkan.

popt adalah parameter-parameter yang telah dioptimalkan.

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, xdata, ydata)
#print the final parameters
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
```

Sekarang kita plot model regresi yang dihasilkan.

*   `curve_fit(sigmoid, xdata, ydata)`: Fungsi ini mencoba menyesuaikan fungsi sigmoid (yang didefinisikan sebelumnya) dengan array xdata dan ydata (yang merupakan versi normal dari data asli). Fungsi ini melakukannya dengan memvariasikan parameter fungsi sigmoid (Beta_1 dan Beta_2 dalam kasus ini) untuk meminimalkan kuadrat perbedaan antara nilai prediksi dan nilai aktual (kriteria kuadrat terkecil).
*   `popt`: Variabel ini akan berisi parameter optimal untuk Beta_1 dan Beta_2 setelah curve_fit selesai. Ini adalah nilai yang paling sesuai dengan fungsi sigmoid dengan titik data yang disediakan.
*   `pcov`: Ini adalah matriks kovarians dari parameter yang ditemukan. Elemen diagonal memberikan varians dari estimasi parameter. Ini dapat digunakan untuk memperkirakan standar deviasi dari estimasi setiap parameter.

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
# split data into train/test
msk = np.random.rand(len(df)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

# buat model dengan train set
popt, pcov = curve_fit(sigmoid, train_x, train_y)

# prediksi dengan test set
y_hat = sigmoid(test_x, *popt)

# metrik evaluasi
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(y_hat , test_y) )
```

## Klasifikasi

Klasifikasi adalah salah satu jenis metode Supervised Learning di mana variabel targetnya berupa kategori atau kelas diskrit. ### Random Forest


Random forest adalah algoritma pembelajaran mesin yang digunakan untuk mengembangkan model prediktif. Ini adalah metode ensemble yang menggabungkan banyak pohon keputusan, di mana setiap pohon dilatih pada subset data yang berbeda. Klasifikasi akhir dibentuk berdasarkan hasil voting dari pohon yang terbentuk. Semakin banyak pohon yang digunakan, semakin baik akurasinya. Random forest digunakan dalam berbagai bidang seperti sistem rekomendasi ahli medis, analisis sentimen, analisis koefisien intensitas material, analisis kinerja spasial dalam bola basket, dan prediksi atrisi karyawan.

Untuk menguraikan lebih lanjut, berikut adalah beberapa poin penting tentang random forest:

- Random forest adalah jenis algoritme pembelajaran terawasi yang dapat digunakan untuk tugas klasifikasi dan regresi.
- Ini adalah metode ensemble yang menggabungkan beberapa pohon keputusan untuk membuat prediksi akhir.
- Setiap pohon keputusan dalam random forest dilatih pada subset data yang berbeda, yang membantu mengurangi overfitting dan meningkatkan akurasi model.
- Random forest adalah algoritma yang kuat yang dapat menangani dataset besar dengan banyak fitur dan dapat digunakan untuk data kategorikal dan kontinu.
- Salah satu keuntungan dari random forest adalah bahwa ia menyediakan ukuran kepentingan fitur, yang dapat digunakan untuk mengidentifikasi variabel yang paling penting dalam model.
- Random forest banyak digunakan di berbagai bidang seperti keuangan, kesehatan, pemasaran, dan olahraga.

Singkatnya, random forest adalah algoritma pembelajaran mesin yang kuat yang dapat digunakan untuk mengembangkan model prediktif di berbagai bidang. Ini adalah metode ensemble yang menggabungkan beberapa pohon keputusan untuk membuat prediksi akhir, dan dikenal karena keakuratan dan kemampuannya untuk menangani set data yang besar.

#### **Berikut adalah bagaimana majority voting bekerja dalam Random Forest:**

1. <b>Bootstrap Aggregating (Bagging)</b> <br>
Random Forest dimulai dengan membuat beberapa pohon keputusan, biasanya ratusan atau ribuan pohon. Setiap pohon dilatih pada subset acak dari data pelatihan, dengan penggantian. Proses ini dikenal sebagai bootstrap aggregating atau bagging. Bagging membantu memperkenalkan keragaman di antara pohon-pohon tersebut, karena setiap pohon melihat subset yang berbeda dari data.<br><br>

2. <b>Pohon Keputusan</b> <br>
Setiap pohon keputusan individu dalam Random Forest dilatih untuk memprediksi variabel target berdasarkan subset acak dari fitur-fitur pada setiap split. Ini lebih meningkatkan keragaman di antara pohon-pohon tersebut.<br><br>

3. <b>Prediksi</b> <br>
Begitu semua pohon keputusan dilatih, mereka dapat digunakan untuk membuat prediksi pada data baru yang belum terlihat. Untuk tugas klasifikasi, setiap pohon memprediksi label kelas dari input, dan untuk tugas regresi, setiap pohon memprediksi nilai numerik.<br><br>

4. <b>Majority Voting (Pemungutan Suara Mayoritas)</b><br>
Setelah semua pohon membuat prediksi, hasil prediksi ini digabungkan dengan cara mayoritas voting. Artinya, kelas atau nilai yang paling sering diprediksi oleh pohon-pohon tersebut diambil sebagai prediksi akhir dari Random Forest.

### K-nearest neighbors (K-NN)

K-nearest neighbors (K-NN) adalah algoritma machine learning yang dapat digunakan untuk tugas klasifikasi dan regresi. Algoritma ini merupakan algoritma non-parametrik, yang berarti bahwa algoritma ini tidak membuat asumsi apa pun tentang distribusi data. Sebaliknya, algoritma ini menggunakan seluruh kumpulan data untuk membuat prediksi.

Berikut adalah beberapa detail tambahan tentang K-NN:
- K-NN bekerja dengan menemukan K tetangga terdekat ke titik data yang diberikan berdasarkan metrik jarak seperti jarak Euclidean atau jarak Manhattan.
- Nilai K adalah hyperparameter yang dapat disetel untuk meningkatkan akurasi model. Nilai K yang lebih kecil akan menghasilkan model yang lebih fleksibel yang lebih rentan terhadap overfitting, sementara nilai K yang lebih besar akan menghasilkan model yang lebih kaku yang lebih rentan terhadap underfitting.
- K-NN dapat digunakan untuk tugas klasifikasi dan regresi. Dalam klasifikasi, outputnya adalah label kelas, sedangkan dalam regresi, outputnya adalah nilai kontinu.
- Salah satu keuntungan dari K-NN adalah algoritme ini merupakan algoritme yang sederhana dan intuitif yang dapat digunakan untuk berbagai macam aplikasi.
- Namun, K-NN dapat menjadi mahal secara komputasi, terutama untuk dataset yang besar, karena memerlukan penghitungan jarak antara setiap titik data dan setiap titik data lainnya dalam dataset.
- K-NN sensitif terhadap pilihan metrik jarak dan penskalaan data. Penting untuk melakukan preprocessing data sebelum menerapkan K-NN untuk memastikan hasilnya akurat.

- Sistem rekomendasi: Meskipun algoritme yang lebih canggih biasanya digunakan, KNN masih dapat diterapkan dalam sistem rekomendasi untuk menyarankan item berdasarkan preferensi pengguna yang serupa.

- Diagnosis medis: KNN dapat membantu dalam mendiagnosis penyakit atau kondisi dengan membandingkan gejala dan riwayat medis pasien dengan kasus serupa.

### Logistic  Regression

Regresi logistik adalah metode analisis statistik yang digunakan untuk memprediksi hasil biner berdasarkan pengamatan sebelumnya terhadap kumpulan data. Ini adalah algoritme klasifikasi yang digunakan untuk menghitung probabilitas terjadinya peristiwa biner dan untuk menangani masalah klasifikasi. Regresi logistik digunakan untuk memprediksi kemungkinan semua jenis hasil "ya" atau "tidak", seperti memprediksi apakah email yang masuk adalah spam atau bukan spam, atau memprediksi apakah transaksi kartu kredit adalah penipuan atau bukan penipuan. Regresi logistik penting karena mengubah perhitungan yang rumit seputar probabilitas menjadi masalah aritmatika yang mudah.

#### Cara Kerja Logistic Regression

1. Model Matematik:

Regresi logistik menggunakan fungsi logistik (atau fungsi sigmoid) untuk memodelkan probabilitas. Fungsi sigmoid memiliki bentuk kurva S dan dapat mengubah setiap nilai input ke dalam rentang antara 0 dan 1, yang bisa diartikan sebagai probabilitas.

sigmoid.jpeg

Di mana z adalah kombinasi linier dari fitur (variabel independen)

image.png

2. Estimasi Koefisien:

Tujuan dari regresi logistik adalah untuk menemukan koefisien yang terbaik, sehingga model dapat memprediksi probabilitas dengan akurat. Ini dilakukan dengan memaksimalkan fungsi log-likelihood dengan teknik seperti "gradient descent".

3. Ambang Batas (Threshold):

Meskipun regresi logistik memberikan probabilitas, dalam banyak kasus kita perlu keputusan kategorikal sebagai hasil akhir (misalnya, spam atau bukan spam). Untuk itu, kita menetapkan ambang batas. Pada klasifikasi biner, ambang batas yang umumnya digunakan adalah 0,5:

- Jika P(Y=1)>0,5, maka prediksi adalah kelas 1.
- Jika P(Y=1)0,5, maka prediksi adalah kelas 0.

```python
iris = load_iris()
X3 = iris.data
y3 = (iris.target == 0).astype(int)  # Convert to binary: 1 if 'setosa', 0 otherwise
```

```python
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train3)
X_test_scaled = scaler.transform(X_test3)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=42)
clf.fit(X_train_scaled, y_train3)
```

```python
# Melakukan prediksi apabila fitur bagian test dimasukkan

y_pred3 = clf.predict(X_test_scaled)
```

```python
accuracy_score(y_test3, y_pred3)
```

```python
confusion_matrix(y_test3, y_pred3)
```

```python
print(classification_report(y_test3, y_pred3))
```
